```python
import numpy as np
import math
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
def simpson(f,h):
  sum = f[0]
  for i in range(1, len(f)-1):
      if i % 2 == 0:
          c = 2
      else:
          c = 4
      sum = sum + (c * f[i])
  sum = sum * (h/3)
  return sum
```

```python
def trapezoidal(f,h):
  sum = f[0]
  for i in range(1, len(f)-1):
    sum = sum + f[i]
  sum = sum * h
  return sum
```

```python
def series(n, x):
  temp = x
  sum = x
  for i in range(2, n + 1):
    temp = ((-1 * temp) * ((x ** 2)) / ((2 * i - 2) * (2 * i - 1)))
    sum = sum + temp
  return sum
```

```python
def calculate(func, t, area, n, method):

  if method == "simpson":
    simp_err = []

    for i in range(3, n, 2):
      x = np.linspace(0, 1, i)
      if(func == 'x'):
        y = x
      elif(func == 'sin(x)'):
        y = np.sin(x)
      elif(func == 'series'):
        x = np.linspace(0, np.pi, i)
        y = [series(t, x) for x in x]
      sum_simp = simpson(y, (1 / (i-1)))
      er = abs(((sum_simp - area) / (area)) * 100)
      simp_err.append(er)

    plt.plot(range(3, n, 2), simp_err, 'r')
    plt.xlabel("N")
    plt.ylabel("Error(%)")
    plt.title("Simpson's rule")
    plt.show()
    print("The exact area is: ",area)
    print("Area by simpson's rule= " + str(sum_simp) + " for iteration = " + str(i-2) + " Error = " + str(((sum_simp - area) / area) * 100))

  elif method == "trapezoidal":
    trap_err = []

    for i in range(2, n):
      x = np.linspace(0, 1, i)
      if(func == 'x'):
        y = x
      elif(func == 'sin(x)'):
        y = np.sin(x)
      elif(func == 'series'):
        x = np.linspace(0, np.pi, i)
        y = [series(t, x) for x in x]
      sum_trap = trapezoidal(y, 1 / (i - 1))
      er = abs(((sum_trap - area) / (area)) * 100)
      trap_err.append(round(er, 4))

    plt.plot(range(2, n), trap_err)
    plt.xlabel("N")
    plt.ylabel("Error(%)")
    plt.title("Trapezoidal rule")
    plt.show()
    print("The exact area is: ",area)
    print("Area by trapezoidal rule= " + str(sum_trap) + " for iteration = " + str(i-2) + " Error = " + str(((sum_trap - area) / area) * 100))
      
```

```python
calculate('x', 0, 0.5, 10, 'simpson')
calculate('x', 0, 0.5, 10, 'trapezoidal')
calculate('sin(x)', 0, -np.cos(1)+1, 6, 'simpson')
calculate('sin(x)', 0, -np.cos(1)+1, 6, 'trapezoidal')
calculate('series', 2,-np.cos(np.pi)+1, 50, 'simpson')
calculate('series', 2,-np.cos(np.pi)+1, 50, 'trapezoidal')
calculate('series', 3,-np.cos(np.pi)+1, 100, 'simpson')
calculate('series', 3,-np.cos(np.pi)+1, 100, 'trapezoidal')
calculate('series', 4,-np.cos(np.pi)+1, 100, 'simpson')
calculate('series', 4,-np.cos(np.pi)+1, 100, 'trapezoidal')

calculate('series', 5,-np.cos(np.pi)+1, 100, 'simpson')
calculate('series', 5,-np.cos(np.pi)+1, 100, 'trapezoidal')

calculate('series', 100,-np.cos(np.pi)+1, 1000, 'simpson')
calculate('series', 100,-np.cos(np.pi)+1, 1000, 'trapezoidal')
```
