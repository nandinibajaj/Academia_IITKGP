Calculate pairwise interaction potential energy as a function of distance r, width dr for 3 interactions

```python
def eval1(ri,rf,dr,sig,eps):
  lam=2
  list1=[]
  x=[]
  y=[]
  for r in np.arange(ri,rf,dr):
    if r<sig:
      V=2
    elif r<(lam*sig) and r>=sig:
      V=-eps
    else:
      V=0
    list1.append([r,V])
    x.append(r)
    y.append(V)
  return (x,y,list1)
```

```python
def eval2(ri,rf,dr,sig,eps,neu):
  list1=[]
  x=[]
  y=[]
  for r in np.arange(ri,rf,dr):
    a=eps*(sig**neu)
    V=a*(r**-neu)
    list1.append([r,V])
    x.append(r)
    y.append(V)
  return (x,y,list1)
```

```python
def eval3(ri,rf,dr,sig,eps):
  list1=[]
  x=[]
  y=[]
  for r in np.arange(ri,rf,dr):
    V=4*eps*(((sig/r)**12)-((sig/r)**6))
    list1.append([r,V])
    x.append(r)
    y.append(V)
  return (x,y,list1)
```

```python
import numpy as np
import pandas as pd

opt = int(input("Which interactions result is wanted:"))
ri=float(input("Enter the value of r initial:"))
rf=float(input("Enter the value of r final:"))
dr=float(input("Enter the value of dr:"))
sig=float(input("Enter the value of sigma:"))
eps=float(input("Enter the value of epsilon:"))

if opt==1:
  x1,y1,sqwell=eval1(ri,rf,dr,sig,eps)
if opt==2:
  x2,y2,softsphere1=eval2(ri,rf,dr,sig,eps,1)
  x3,y3,softsphere2=eval2(ri,rf,dr,sig,eps,10)
if opt==3:
  x4,y4,lj=eval3(ri,rf,dr,sig,eps)
  
```
